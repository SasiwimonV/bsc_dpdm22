##ใช้ MAPE

```python
est2['ประเมิน1'] = est2['post id']. map(est1.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
est2
```

```python
est2 = est2.rename(columns={"ประเมินราคาซื้อ-ขาย": "ประเมิน2"})
est2
```

```python
table = est2[['post id','ราคาทรัพย์','ประเมิน1','ประเมิน2']]
```

```python
table = table.dropna(subset=['ราคาทรัพย์']) # row หายไป 29 row เพราะ มีค่า Nan เราไม่ fillna(0) เพราะ เวลาแทนสูตรจะหาไม่ได้
```

```python
table
```

```python
table1 = table['ประเมิน1']
```

```python
table2 = table['ประเมิน2']
```

```python
table3 = table['ราคาทรัพย์']
```

```python
table.isnull().any() #check
```

```python
import numpy as np #numpy
```

```python
def mape(y_test, pred):
    y_test, pred = np.array(y_test), np.array(pred)
    mape = np.mean(np.abs((y_test - pred) / y_test))
    return mape
```

```python
mapetable = mape(table1, table2)
mapetable # ค่า mape ของผู้ประเมิน1 กับผู้ประเมิน 2
print(f'ผู้ประเมิน 1 กับ ผู้ประเมิน 2 มีค่า MAPE อยู่ที่', mapetable)
```

```python
mapetable2 = mape(table1, table3)
mapetable2 # ค่า mape ของผู้ประเมิน1 กับ ราคาทรัพย์
print(f'ผู้ประเมิน 1 กับ ราคาทรัพย์ มีค่า MAPE อยู่ที่', mapetable2)
```

```python
mapetable3 = mape(table2, table3)
mapetable3
print(f'ผู้ประเมิน 2 กับ ราคาทรัพย์ มีค่า MAPE อยู่ที่', mapetable3)
```

```python
mapetable4 = mape(table2, table1)
mapetable4
print(f'ผู้ประเมิน 2 กับ ผู้ประเมิน 1 มีค่า MAPE อยู่ที่', mapetable4)
```

```python
mapetable5 = mape(table3, table1)
mapetable5
print(f'ราคาทรัพย์ กับ ผู้ประเมิน 1 มีค่า MAPE อยู่ที่', mapetable5)
```

```python
mapetable6 = mape(table3, table2)
mapetable6
print(f'ราคาทรัพย์ กับ ผู้ประเมิน 2 มีค่า MAPE อยู่ที่', mapetable6)
```
